```python
from google import google
from lxml import etree
from pyhanlp import *
import pandas as pd
import numpy as np
import requests
import smtplib 
from email.mime.multipart import MIMEMultipart 
from email.mime.text import MIMEText 
from email.mime.base import MIMEBase 
from email import encoders 
#query = ["洗錢", "賄賂", "貪污", "詐欺", "走私", "犯罪", "內線交易", "市場操縱", "毒品"]
query = ["內線交易", "賄賂"]#, "貪污", "詐欺", "走私", "犯罪", "內線交易", "市場操縱", "毒品"]
#newspaper = ["聯合報", "經濟日報", "台灣蘋果日報", "中國時報", "自由時報", "工商時報", "聯合晚報"]
newspaper = ["聯合報","經濟日報"]
num_page = 1
address = ["https://udn.com/news/story", "https://money.udn.com/money/story"]
hanlp = JClass('com.hankcs.hanlp.HanLP')
Segment = hanlp.newSegment()
segment = HanLP.newSegment().enableNameRecognize(True).enableTranslatedNameRecognize(True);
NLPTokenizer = JClass('com.hankcs.hanlp.tokenizer.NLPTokenizer')
TextRankSentence = JClass("com.hankcs.hanlp.summary.TextRankSentence")
column=['人名','查詢關鍵字','新聞出處','連結','摘要']
temp=[]
for q in query:
    for n in newspaper:
        num_page = 1
        search_results = google.search(q+" "+n+" 新聞", num_page)        
        #print("="*30 + q + "="*30)
        for ele in search_results:
            if any(target in ele.link for target in address):
#            if "https://udn.com/news/story" or "https://money.udn.com/money/story" in ele.link:                
                #print(ele.link)
                html = requests.get(ele.link)
                etree_structure = etree.HTML(html.text)
                text = etree_structure.xpath(r'//*[@id="story_body_content"]//p//text()')
                #r'//*[@id="story_body_content"]//p/text()'
                article = " ".join(text)
                article = article.strip()
                abstract = HanLP.extractSummary(article, 5)                
                #print("="*10, "Person", "="*10)
                words = segment.seg(article)
                for word in words:
                    #print(word)
                    if "nr" in str(word.nature):
                        temp.append([word.word,q,n,ele.link,abstract])                        
                        #print(word.word)
        #print("="*60 + "\n")
suspectors_data = pd.DataFrame(temp,columns=column)
suspectors_data
```
