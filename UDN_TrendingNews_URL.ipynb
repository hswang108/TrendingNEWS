```python
from lxml import etree
import requests

root = "https://udn.com/rank/pv/2/6644/"
address = root+str(1)
html = requests.get(address)
etree_structure = etree.HTML(html.text)
last_page = etree_structure.xpath(r'//*[@id="ranking"]/div[3]/gonext/a[2]/@href')
last_page = " ".join(last_page).split("/6644/")[1]
print("There are",last_page,"pages")


for end in range(int(last_page)):
    address = root+str(end+1)
    print("page", end+1, ":")
    i = 0
    space = 0
    while space < 3:
        html = requests.get(address)
        etree_structure = etree.HTML(html.text)
        text = etree_structure.xpath(r'//*[@id="breaknews_body"]/dl/dt['+str(i+1)+']/h2/a/@href')
        i = i+1
        if not text:
            space = space + 1
        else:
            print(" ".join(text))
            space = 0
        
```
